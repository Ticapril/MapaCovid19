```python
from matplotlib import pyplot as plt


x = []
y = []

dataset = open('dados.txt','r')
for linha in dataset:
    linha = linha.strip()
    A,  B = linha.split(';')
    x.append(int (A))
    y.append(int (B))
    
    dataset.close()
    
    plt.plot(x,y)
    plt.show()
```

```python
import matplotlib.pyplot as plt
#titulo do gráfico
x = [1,2,3,4]
y = [1,4,2,3]

plt.plot(x,y)
plt.title("Meu Grafico")
plt.show()
```

```python
import matplotlib.pyplot as plt
#titulo do gráfico
x = [1,2,3,4]
y = [1,4,2,3]
a = [5,6,8]
b = [8,5,6]

plt.plot(a,b, color = 'r', linestyle = 'solid', label = "Grafico 1")

plt.plot(x,y, color = 'b', linestyle = 'dashed', label = 'Grafico 2')
plt.legend()
plt.title("Meu Grafico")
plt.xlabel("Titulo do eixo X")
plt.ylabel("Titulo do eixo Y")
plt.show()
```
