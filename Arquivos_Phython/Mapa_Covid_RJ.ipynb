```python
##primeira célula é feito a importação das bibliotecas
import pandas as pd 
import numpy as np

import folium
from folium import plugins
#importando as bibliotecas o folium trabalha com camadas de mapas, mapa gerado em svg
```

```python
##configuração do mapa (map set up) mapa RJ
mapa  = folium.Map(
                   width = "100%",
                   height = "100%",
                   location = [-22.9035,-43.2096],
                   zoom_start = 8
                   )
```

```python
mapa
```

```python
df = pd.read_csv('brasil-io-covid-19RJ.csv')
```

```python
df
```

```python
estados = df.loc [df.place_type == 'state' , ['state','confirmed', 'deaths', 'is_last']]
```

```python
estados.head()
```

```python
estados.isnull().sum()
```

```python
ultimas = estados.loc[estados.is_last == True, ['state', 'confirmed', 'deaths']]    
```

```python
len(ultimas)
```

```python
ultimas.head()
```

```python
ultimas['log_confirmed'] = np.log(ultimas['confirmed']+1)
ultimas['log_deaths'] = np.log(ultimas['deaths']+1)
```

```python
import json

br_estados = "geojs-33-mun.json"
geo_json_data = json.load(open(br_estados))
```

```python
geo_json_data
```

```python
mapa.choropleth(
    geo_data = geo_json_data,
    name = "CASOS",
    data = ultimas,
    columns = ['state','log_confirmed'],
    key_on = 'feature.id',
    fill_color = "Reds",
    fill_opacity = 0.8,
    line_color = 'white',
    line_opacity = 0.8,
    show = True,
    legend_name = "LOG DE CASOS DO NOVO COVID-19 NO RJ"



)
```
