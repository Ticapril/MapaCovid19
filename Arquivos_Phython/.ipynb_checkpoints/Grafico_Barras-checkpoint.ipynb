```python
from matplotlib import pyplot as plt

x = [1,2,3,4,5,6]
y = [5,9,7,3,8,2]

plt.bar(x,y)

plt.title("Gr√°fico de Barras")
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")

plt.show()
```

```python
##compara√ß√£o de Gr√°ficos de Barra

from matplotlib import pyplot as plt

x1 = [1,3,5,7,9,11]
y1 = [5,9,7,3,8,2]

x2 = [2,4,6,8,10,12]
y2 = [8,5,2,3,9,7]

## configurando o gr√°fico cor etiqueta

plt.bar(x1,y1, color = 'g', label = "Barra 1")
plt.bar(x2,y2, color = 'r', label = "Barra 2")

plt.title("Compara√ß√£o de Gr√°ficos de Barras")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.legend()
plt.show()
```

```python
##plotar gr√°fico compara√ß√£o barrasxlinhas

from matplotlib import pyplot as plt

x = [1,3,5,7,9,11]
y = [5,9,7,3,8,2]

## set up

plt.plot(x,y, color = 'r', label = "Linha")
plt.bar(x,y, color = 'b', label = "Barra")

plt.title("Compara√ß√£o de Gr√°ficos  de Barras")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.legend()
plt.show()
```

```python
## Plotar um Gr√°fico de Barra Horizontal

plt.barh(x,y)
from matplotlib import pyplot as plt

dados = [ 1,3,5,7,9,11]
nomes = ['RJ', 'SP', 'RS', 'MG', 'MA', 'CE']

plt.barh(nomes,dados, color = 'g', label = "Barra")

plt.title("Gr√°fico de Barra Horizontal")
plt.xlabel("Eixo X")
plt.ylabel("Eixo y")
plt.legend()

plt.show()


```
